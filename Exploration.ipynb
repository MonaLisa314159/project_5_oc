## 1. Nettoyage des tags :

```python
import re

# Fonction pour nettoyer les tags
def clean_tags(tag_string):
    """ clean Tags : extract tags between brackets using regular expression"""
    
    tags = re.findall(r'<(.*?)>', tag_string)
    return tags

# Appliquer la fonction de nettoyage des tags à la colonne 'Tags'
df['Tags'] = df['Tags'].apply(clean_tags)

# Afficher les premières lignes du DataFrame après le nettoyage des tags
df.head(5)
```

```python
def print_sample(df, col1, col2, n_sample):
    """ print out n_sample documents """
    samp = df.sample(n_sample)
    
    for idx, ser in samp.iterrows():
        tag = f" Tags : {ser['Tags']}"
        print(tag, "\n")
        print(f" {col1, ser[col1]} \n")
        print(f"{col2, ser[col2]} \n")
```

```python
print_sample(df, 'Title', 'Body', 5)
```

## 2. Nettoyage des textes :

```python
def extract_code(text):
    """ extract code from Body : using regular expression  
       code between the <pre><code> tags """
    
    code_blocks = re.findall(r'<pre><code>(.*?)<\/code><\/pre>', text, flags=re.DOTALL)
    return code_blocks
```

```python
def remove_code(text):
    """ remove extracted code from Body """
    
    text_without_code = re.sub(r'<pre><code>.*?<\/code><\/pre>', '', text, flags=re.DOTALL)
    return text_without_code
```

```python
# Appliquer la fonction extract_code à la colonne "Body"
df['Extracted_code'] = df['Body'].apply(extract_code)

# Appliquer la fonction remove_code à la colonne "Body"
df['Body'] = df['Body'].apply(remove_code)

# Appliquer la fonction remove_code à la colonne "Corpus"
df['Corpus'] = df['Corpus'].apply(remove_code)
```

```python
from bs4 import BeautifulSoup
import warnings
warnings.filterwarnings("ignore")

def clean_text(text):
    """Clean text : remove HTML tags and other non-text elements"""
    soup = BeautifulSoup(text, 'html.parser')
    cleaned_text = soup.get_text(separator=' ')
    return cleaned_text
```

```python
# Appliquer la fonction clean_text aux colonnes "Body" et "Corpus"
df['Body'] = df['Body'].apply(clean_text)
df['Corpus'] = df['Corpus'].apply(clean_text)
```

```python
print_sample(df, 'Body', 'Corpus', 2)
```

```python
def normalize_text(text):
    """ Clean text : 
    Lower text
    Replace all occurrences of "c++" whith cplusplus, "c#" with "csharp", 
    "\.net" with "dotnet and "d3\.js" with "d3js" 
    replace all characters that are not letters with a space """
    
    text = text.lower()
    text = re.sub("c\+\+","cplusplus", text)
    text = re.sub("c#","csharp", text)
    text = re.sub("\.net","dotnet", text)
    text = re.sub("d3\.js","d3js", text)
    text = re.sub("[^a-zA-Z]"," ", text)
    return text
```

```python
# Appliquer la fonction normalize_text aux colonnes "Body" et "Corpus"
df['Body'] = df['Body'].apply(normalize_text)
df['Corpus'] = df['Corpus'].apply(normalize_text)
```

```python
def tokenizer_fct(text):
    tokenizer = nltk.RegexpTokenizer(r'\w+')
    word_tokens = tokenizer.tokenize(text)
    return word_tokens

def remove_stop_word(list_words):
    """ Remove stop words from text """
    #     stop_w = set(stopwords.words('english'))
    stop_w = list(set(stopwords.words('english'))) + ['[', ']', ',', '.', ':', '?', '(', ')']
    filtered_w = [w for w in list_words if not w in stop_w]
    filtered_w2 = [w for w in filtered_w if len(w) > 2]
    return filtered_w2

def lemm_fct(text):
    """ Lemmatization """
    lemmatizer = WordNetLemmatizer()
    lem_w = [lemmatizer.lemmatize(w) for w in text]
    return lem_w
```

```python
# Tokenisation des colonnes 'Title', 'Body' et 'Corpus':
df_2['Title_tokens'] = df_2['Title'].apply(tokenizer_fct)
df_2['Body_tokens'] = df_2['Body'].apply(tokenizer_fct)
df_2['Corpus_tokens'] = df_2['Corpus'].apply(tokenizer_fct)

df_2.head(5)
```

```python
# Suppression des stop words
df_2['Title_tokens'] = df_2['Title_tokens'].apply(remove_stop_word)
df_2['Body_tokens'] = df_2['Body_tokens'].apply(remove_stop_word)
```
